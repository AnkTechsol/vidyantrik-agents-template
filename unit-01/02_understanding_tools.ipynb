In the previous notebook, we chatted with an agent. Now, we will give it **tools** to interact with the world.

## 1. What are Tools?

Tools are functions that an agent can call. They allow the agent to perform actions like:
-   Searching the web
-   Calculating numbers
-   Generating images

In `smolagents`, a tool is just a Python function decorated with `@tool`.

```python
import sys
import os

# Add the current directory to path so we can import the tool
sys.path.append(os.getcwd())
```

## 2. Using a Custom Tool

We have prepared a simple calculator tool in `tool_calculator.py`. Let's import it.

```python
from tool_calculator import calculator

print("Tool loaded:", calculator.name)
print("Description:", calculator.description)
```

## 3. Creating an Agent with Tools

Now we create a `CodeAgent` and pass our `calculator` tool to it.

```python
from smolagents import CodeAgent, HfApiModel

# Initialize the model
model = HfApiModel(model_id="Qwen/Qwen2.5-Coder-32B-Instruct")

# Create the agent with the calculator tool
agent = CodeAgent(tools=[calculator], model=model)

# Run the agent with a math problem
agent.run("What is 2024 multiplied by 5?")
```

## 4. Complex Reasoning

Let's see if the agent can handle a multi-step problem.

```python
agent.run("Subtract 50 from 100, then divide the result by 2.")
```
